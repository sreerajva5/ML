## recommendation engine

Helps to create recommendation for users from available data.

Creating a movie recommentation with a movie rating data as per user rating. Available fields in the dataset are:- user_id, movie_id, rating, time_stamp

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'D:\E\Courses\Edureka\Recommend.csv', names=['user_id', 'movie_id', 'rating', 'time_stamp'])
```

```python
df.head(1)
```

```python
df.shape
```

```python
n_users = df.user_id.unique().shape[0]
n_users
```

```python
n_movies = df.movie_id.unique().shape[0]
n_movies
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(df, test_size=0.25)
```

```python
train_data_matrix = np.zeros((n_users, n_movies))
```

```python
pd.DataFrame(train_data_matrix)
```

```python
list(train_data.itertuples())
```

```python
for i in train_data.itertuples():
    train_data_matrix[i[1]-1, i[2]-1] = i[3]
```

```python
pd.DataFrame(train_data_matrix)
```

```python
test_data_matrix = np.zeros((n_users, n_movies))
```

```python
for i in test_data.itertuples():
    test_data_matrix[i[1]-1, i[2]-1] = i[3]
```

```python
pd.DataFrame(test_data_matrix)
```

```python
from sklearn.metrics import pairwise_distances
```

```python
user_similarity = pairwise_distances(train_data_matrix, metric='cosine')
mean_user_rating = train_data_matrix.mean(axis=1)[:,np.newaxis]
```

```python
pd.DataFrame(user_similarity)
```

```python
mean_user_rating
```

```python
rating_diff = train_data_matrix - mean_user_rating
```

```python
pd.DataFrame(rating_diff)
```

```python
user_pred = mean_user_rating + user_similarity.dot(rating_diff)/np.array([np.abs(user_similarity).sum(axis=1)]).T
```

```python
pd.DataFrame(user_pred)
```

The higher numbers represents more like and lower number represents dislike or lesser like towards the movie.
