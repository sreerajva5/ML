##### decomposition of time series

```python
# we can treat a time series as a combination of 4 parts (level, trend, seasonality, noice)

# Addictive model : Level + trend + seasonality + noice

# Multiplicative model : Level * trend * seasonality * noice

# decomposition : process of identifying these 4 parts from a time series
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
import pandas as pd
df = pd.read_csv(r'C:\E\E_drive_copy\sample_data\ts_ud\Resources\Data + Code\us-airlines-monthly-aircraft-miles-flown.csv', parse_dates=[0])
```

```python
df.head()
```

```python
df.index = df['Month']

df.head()
```

```python
# checking level, trend, seasonality, noice with addictive model

df_sd_add = seasonal_decompose(df['MilesMM'], model='additive')
```

```python
df_sd_add.plot()
```

```python
df_sd_mul = seasonal_decompose(df['MilesMM'], model='multiplicative')
```
