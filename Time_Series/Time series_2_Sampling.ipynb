```python
# Resampling: Changing the frequenncy of available data to match the frequency of required forecast.
# (if frequency of available data is not matching with frequency of the data tobe forecasted)

# There are 2 types of resampling

# Upsampling     :  Add more samples to data and increase the frequency (interpolation technique)

# Downsampling   :  Remove samples from existing data and reduce the frequency (aggregated values)

# Another reason for resampling can be feature engineering. Indert more features into data.
```

```python
df = pd.read_csv(r'C:\E\E_drive_copy\sample_data\ts_ud\Resources\Data + Code\us-airlines-monthly-aircraft-miles-flown.csv', parse_dates=[0])

df
```

##### downsampling

```python
quarterly_mean_df = df.resample('Q', on='Month').mean()

# using mean of 3 months calculated the quarterly miles

quarterly_mean_df
```

```python
quarterly_sum_df = df.resample('Q', on='Month').sum()

# using sum of 3 months miles calculated the quarterly miles

quarterly_sum_df
```

```python
quarterly_max_df = df.resample('Q', on='Month').max()

# using max of 3 months miles calculated the quarterly miles

quarterly_max_df
```

```python
quarterly_min_df = df.resample('Q', on='Month').min()

# using min of 3 months miles calculated the quarterly miles

quarterly_min_df
```

```python
yearly_sum_df = df.resample('A', on='Month').sum()

# using sum of 12 months miles calculated the yearly miles

yearly_sum_df
```

##### upsampling

```python
daily_mean_df = df.resample('D', on='Month').mean()

# try to convert to daily miles from available monthly stats (need to interpolate in next step)

daily_mean_df
```

```python
daily_mean_df_interpolated = daily_mean_df.interpolate(method='linear')

# linear method equally divides the difference between 2 neares values into required number of values
# difference between 2 nearest values will be equal

daily_mean_df_interpolated
```

```python
daily_mean_df_interpolated.plot()
```

```python
daily_mean_df_interpolated_2 = daily_mean_df.interpolate(method='spline', order=2)

# using second degree polynomial to perform upsampling

daily_mean_df_interpolated_2
```

```python
daily_mean_df_interpolated_2.plot()

# sharp edges are minimal compared to linear method in polynomial method
```
