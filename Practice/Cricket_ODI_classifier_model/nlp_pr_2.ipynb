```python
# punctuations remove

f = open(r'C:\E\E_drive_copy\sample_data\nlp\ek\ast\data\Macbeth-ArticlesRemoved.txt', 'r')
txt = f.read()
txt_2 = ''
for i in txt:
    if i not in '.,:;#()*^%$\'"-+=&?<>[]{}@':
        txt_2 += i
```

```python
print(txt_2)
```

```python
f = open(r'C:\E\E_drive_copy\sample_data\nlp\ek\ast\data\Macbeth-punctuationsRemoved.txt', 'w')
f.write(txt_2.strip())
f.close()
```

```python
# number of words
from nltk.corpus import names

len(names.words('male.txt')) + len(names.words('female.txt'))
```

```python
names_list = names.words('male.txt') + names.words('female.txt')

print(names_list)
```

```python
# count of names for each alphabet

count_dict = {}

for i in names_list:
    if i[0] not in count_dict:
        count_dict[i[0]] = 1
    else:
        count_dict[i[0]] += 1

print(count_dict)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.bar(count_dict.values(), height=count_dict.keys())
plt.xlabel('frequency of names')
plt.ylabel('alphabets')
plt.savefig(r'C:\E\E_drive_copy\sample_data\nlp\ek\ast\data\FrequencyOfNamesOfEachAlphabet.png')
```
