```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
df = df[['Team_1', 'Team_2', 'Ground', 'Host_Country', 'Venue_Team1', 'Innings_Team1', 'Winner']]
```

```python
df2 = df.copy()
```

```python
df2.head(1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in df.columns:
    df[i] = le.fit_transform(df[i])
```

```python
df.shape
```

```python
x = df.iloc[:, :6].values
y = df.iloc[:, 6].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x , y, random_state=1, test_size=0.25)
```

```python
len(x_train), len(x_test), len(y_train), len(y_test)
```

#### xgboost

```python
!pip install xgboost
```

```python
from xgboost import XGBClassifier

xgbc = XGBClassifier()
xgbc.fit(x_train, y_train)
```

```python
y_pred_xgbc = xgbc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
accuracy_score(y_test, y_pred_xgbc)
```

```python
#india vs australia match on 29th Nov predition
xgbc.predict(np.array([[1,8,140,0,1,0]]))
```

```python
# south africa vs england match on 29th Nov predition (exact venue not available)
xgbc.predict(np.array([[17,6,17,14,1,0]]))
```

```python
from xgboost import XGBRFClassifier
xgbrf = XGBRFClassifier(num_parallel_tree=20)
xgbrf.fit(x_train, y_train)
```

```python
y_pred_xgbrf = xgbrf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_xgbrf)
```

```python
#india vs australia match on 29th Nov predition
xgbrf.predict(np.array([[1,8,140,0,1,0]]))
```

```python
# south africa vs england match on 29th Nov predition (exact venue not available)
xgbrf.predict(np.array([[17,6,17,14,1,0]]))
```

#### trying grid search over xgbrf

```python
from sklearn.model_selection import GridSearchCV
param = [{'num_parallel_tree':list(range(2,101)), 'importance_type':["gain", "weight", "cover", "total_gain", "total_cover"]}]
```

```python
grid_search = GridSearchCV(estimator=xgbrf, param_grid=param, n_jobs=-1, scoring='accuracy')
grid_search = grid_search.fit(x_train, y_train)
```

```python
best_score = grid_search.best_score_
best_param = grid_search.best_params_
```

```python
best_score, best_param
```
