# Grid Search Algorithm

####### Grid search algorithm will help to identify which parameters are best to apply to predict with other algorithms.

Trying to predict if customer will buy the product or not using different classification algorithm and then applyting grid search algorithm to know which parameters are best to predict.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'D:\Courses\Edureka\Social_Network_Ads.csv')
```

```python
print(df.shape)
df.head(2)
```

```python
x = df.iloc[:, [2,3]].values
```

```python
x
```

```python
# as we apply KNN classification algorithm need to apply standardscalar to x (to make both independant variables in same scale)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x
```

```python
y = df.iloc[:,4].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=99)
```

####### applying support vector machine classification algorithm to predict if customer will purchase product

```python
from sklearn.svm import SVC
classifier_svm = SVC(kernel='sigmoid')
classifier_svm.fit(x_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier()
classifier_knn.fit(x_train, y_train)
```

```python
y_predict_svm = classifier_svm.predict(x_test)
```

```python
y_predict_knn = classifier_knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print(confusion_matrix(y_test, y_predict_svm))
accuracy_score(y_test, y_predict_svm)
```

```python
print(confusion_matrix(y_test, y_predict_knn))
accuracy_score(y_test, y_predict_knn)
```

###### applying grid search algorithm on SVM to check which parameter can give better accuracy score

```python
from sklearn.model_selection import GridSearchCV

parameters = [{'C':[2,3,4,5,6,7], 'kernel':['sigmoid', 'rbf', 'linear']}]  # grid search will check for these parameters
```

```python
grid_search = GridSearchCV(estimator=classifier_svm, param_grid=parameters,
                          scoring='accuracy', n_jobs=-1)

# estimator - checking for which algorithm, parameter_grid - for which parameters we have to check best accuracy
# scoring - accuracy score, n_jobs=-1 means entire CPU dedicating forthis process

grid_search = grid_search.fit(x_train, y_train)
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_
```

```python
best_accuracy
```

```python
best_parameters
```

```python
# best is with 4 is the best number and kernel is rbf
```

###### applying grid search algorithm on KNN to check which parameter can give better accuracy score

```python
from sklearn.model_selection import GridSearchCV

parameters = [{'n_neighbors':lst}]
```

```python
grid_search = GridSearchCV(estimator=classifier_knn, param_grid=parameters, scoring='accuracy', n_jobs=-1)
grid_search = grid_search.fit(x_train, y_train)
```
