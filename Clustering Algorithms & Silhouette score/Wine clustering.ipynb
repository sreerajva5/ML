```python
x = df.values
```

```python
x
```

```python
# normalizing

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x = ss.fit_transform(x)
```

```python
x
```

```python
from sklearn.cluster import KMeans
elbow = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init='k-means++')
    kmeans.fit(x)
    elbow.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), elbow)
```

```python
# K-Means clustering with 3 clusters

kmeans = KMeans(n_clusters=3, init='k-means++')
labels = kmeans.fit_predict(x)
```

```python
labels
```

```python
pd.concat([df, pd.DataFrame(labels)], axis=1)
```

```python
plt.scatter(x[labels==0, 0], x[labels==0, 1], s=80, c='r')
plt.scatter(x[labels==1, 0], x[labels==1, 1], s=80, c='b')
plt.scatter(x[labels==2, 0], x[labels==2, 1], s=80, c='k')
```
