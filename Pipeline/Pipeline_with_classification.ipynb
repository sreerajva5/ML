#### classification models with pipeline

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline

pipe_lr = Pipeline([('scaler1', MinMaxScaler()), ('lr', LogisticRegression())])
pipe_dt = Pipeline([('scaler1', MinMaxScaler()), ('dt', DecisionTreeClassifier())])
pipe_rf = Pipeline([('scaler1', MinMaxScaler()), ('rf', RandomForestClassifier())])
pipe_sv = Pipeline([('scaler1', MinMaxScaler()), ('sv', SVC())])
pipe_nb = Pipeline([('scaler1', MinMaxScaler()), ('sv', GaussianNB())])
pipe_ada = Pipeline([('scaler1', MinMaxScaler()), ('sv',  AdaBoostClassifier())])
pipe_knn = Pipeline([('scaler1', MinMaxScaler()), ('sv', KNeighborsClassifier())])
pipe_xgb = Pipeline([('scaler1', MinMaxScaler()), ('sv', XGBClassifier())])
```

```python
pipe_lr2 = Pipeline([('scaler2', StandardScaler()), ('lr', LogisticRegression())])
pipe_dt2 = Pipeline([('scaler2', StandardScaler()), ('dt', DecisionTreeClassifier())])
pipe_rf2 = Pipeline([('scaler2', StandardScaler()), ('rf', RandomForestClassifier())])
pipe_sv2 = Pipeline([('scaler2', StandardScaler()), ('sv', SVC())])
pipe_nb2 = Pipeline([('scaler2', StandardScaler()), ('sv', GaussianNB())])
pipe_ada2 = Pipeline([('scaler2', StandardScaler()), ('sv',  AdaBoostClassifier())])
pipe_knn2 = Pipeline([('scaler2', StandardScaler()), ('sv', KNeighborsClassifier())])
pipe_xgb2 = Pipeline([('scaler2', StandardScaler()), ('sv', XGBClassifier())])
```

```python
pipe_lr3 = Pipeline([('scaler3', RobustScaler()), ('lr', LogisticRegression())])
pipe_dt3 = Pipeline([('scaler3', RobustScaler()), ('dt', DecisionTreeClassifier())])
pipe_rf3 = Pipeline([('scaler3', RobustScaler()), ('rf', RandomForestClassifier())])
pipe_sv3 = Pipeline([('scaler3', RobustScaler()), ('sv', SVC())])
pipe_nb3 = Pipeline([('scaler3', RobustScaler()), ('sv', GaussianNB())])
pipe_ada3 = Pipeline([('scaler3', RobustScaler()), ('sv',  AdaBoostClassifier())])
pipe_knn3 = Pipeline([('scaler3', RobustScaler()), ('sv', KNeighborsClassifier())])
pipe_xgb3 = Pipeline([('scaler3', RobustScaler()), ('sv', XGBClassifier())])
```

```python
pipe_line = [pipe_lr, pipe_dt, pipe_rf, pipe_sv, pipe_nb, pipe_ada, pipe_knn, pipe_xgb, pipe_lr2, pipe_dt2, pipe_rf2, pipe_sv2, pipe_nb2, pipe_ada2, pipe_knn2, pipe_xgb2, pipe_lr3, pipe_dt3, pipe_rf3, pipe_sv3, pipe_nb3, pipe_ada3, pipe_knn3, pipe_xgb3]
```

```python
models = {0 : 'logistic_regression_minmax', 1 : 'decision_tree_minmax', 2 : 'random_forest_minmax', 3 : 
          'svc_minmax', 4 : 'naive_bayes_minmax', 5 : 'ada_minmax', 6 : 'knn_minmax', 
          7 : 'xgb_minmax', 8 : 'logistic_regression_standardscaler', 9 : 'decision_tree_standardscaler', 
          10 : 'random_forest_standardscaler', 11 : 'svc_standardscaler', 12 : 'naive_bayes_standardscaler', 
          13 : 'ada_standardscaler', 14 : 'knn_standardscaler', 15 : 'xgb_standardscaler', 
          16 : 'logistic_regression_robustscaler', 17 : 'decision_tree_robustscaler', 18 : 'random_forest_robustscaler', 
          19 : 'svc_robustscaler', 20 : 'naive_bayes_robustscaler', 21 : 'ada_robustscaler', 22 : 'knn_robustscaler', 
          23 : 'xgb_robustscaler'}
```

```python
for i in pipe_line:
    i.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score
```

```python
scores_dict = {'model':list(models.values()), 'accuracy':[], 'precision':[], 'f1_score':[]}
```

```python
print(scores_dict)
```

```python
for i,j in enumerate(pipe_line):
    y_pred = j.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    scores_dict['accuracy'].append(acc)
    pr = precision_score(y_test, y_pred, average='micro')
    scores_dict['precision'].append(pr)
    f1 = f1_score(y_test, y_pred, average='micro')
    scores_dict['f1_score'].append(f1)
```
